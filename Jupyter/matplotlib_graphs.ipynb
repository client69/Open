```python
#  CS20125_AMARDEEP BHAGAT
#   DSC LAB3




#here i have taken three different dataset 
#  1st dataset is placement.csv which contains placement statistic and it have 9 rows and two column 
#  and used for ploting bar graph

# 2nd and 3rd dataset is created using list and is small dataset and used for scatter plot and pie plot

#  4th dataset is large dataset, file name is "nirf.csv" it contains the nirf details of engineering college
#  and dataset is taken from kaggle it has 200 rows and 12 columns


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style

style.use('ggplot') #styling of graph

url = "placement.csv" #url of csv file is stored in url variable
file = pd.read_csv(url) #import csv file in file
df = pd.DataFrame(file,columns = ['Year','percentage']) #we can also use dataframe to plot bar graph
print(df)
plt.bar(file["Year"],file["percentage"],width = 0.5,color = "red") #plot bar graph
plt.xlabel("year")  # x label  
plt.ylabel("placement percentage") #y-axis label
plt.title("NITW placement Mtech CSE statistics")
plt.legend(["Placement"],loc = "upper left") #to show the meaning of colored bar
plt.show()


```

```python
style.use('ggplot')
x = [2016,2017,2018,2019,2020]
y = [40,60,50,80,99]
plt.scatter(x,y,s=100,c="blue",marker = "o")  #marker and color defined for graph
plt.xlabel("year")  #label x-axis
plt.ylabel("placement percentage") #label y-axis
plt.title("NITW placement statistics") #title of graph
plt.xlim()
plt.show()
```

```python
style.use('ggplot')
Branch = ['CSE', 'ME', 'CIVIL', 'EE']
sizes = [50,30,10,10]
explode = (0, 0.1, 0, 0)  # it is used to show the data little apart

fig1, ax1 = plt.subplots() #subplot
ax1.pie(sizes, explode=explode, labels=Branch, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title("Branch statistics")
plt.show()
```

```python
file2 = pd.read_csv("nirf.csv")
df1 = pd.DataFrame(file2)
print(df1)
plt.xlabel("x axis")
plt.plot(df1["score"],df1["oi"]) #plot
plt.show()
```

```python
S=df1["score"]
T=df1["oi"]
R=df1["rpc"]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.hist([S,T,R],rwidth = 1, color = ['green','orange','red'],label = ['score','oi','rpc']) #histogram plot
plt.legend()
plt.show()
```
